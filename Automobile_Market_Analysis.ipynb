## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Replacing '?' with NaN (I assume this dataset uses '?' for missing)
df.replace('?', np.nan, inplace=True)

print("Missing values in each column:")
print(df.isnull().sum())
````

````python
# Converting object-based numeric columns to proper float
numeric_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
````

````python
# Mapping number of cylinders to numeric
cylinder_map = {
    'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'eight': 8, 'twelve': 12
}
df['num-of-cylinders'] = df['num-of-cylinders'].map(cylinder_map)

# Also map number of doors
door_map = {'two': 2, 'four': 4}
df['num-of-doors'] = df['num-of-doors'].map(door_map)
````

````python
# Filling missing values for numerical features with mean
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        print(f"Filling missing values in {col} with mean.")
        df[col].fillna(df[col].mean(), inplace=True)
````

````python
# Filling missing values in 'num-of-doors' with mode
if df['num-of-doors'].isnull().sum() > 0:
    print("Filling 'num-of-doors' with mode.")
    df['num-of-doors'].fillna(df['num-of-doors'].mode()[0], inplace=True)
````

````python
# Convert relevant columns to 'category' dtype
cat_cols = ['make', 'fuel-type', 'aspiration',
            'body-style', 'drive-wheels', 'engine-location',
            'engine-type', 'fuel-system']

for col in cat_cols:
    df[col] = df[col].astype('category')
````

````python
# Convert symboling to risk category.
def map_symboling(symbol):
    if symbol >= 1:
        return 'High Risk'
    elif symbol == 0:
        return 'Moderate Risk'
    else:
        return 'Low Risk'

df['risk-category'] = df['symboling'].apply(map_symboling)
df['risk-category'] = df['risk-category'].astype('category')

print("Risk category distribution:")
print(df['risk-category'].value_counts())
````

````python
# Analyzing symboling vs price to understand risk ratings and pricing
print("Average price per symboling risk value:")
print(df.groupby('symboling')['price'].mean())
````

````python
# Cars with turbo engines
turbo_cars = df[df['aspiration'] == 'turbo']
print("Number of turbo engine cars:", turbo_cars.shape[0])

# Cars with rear engine (rare)
rear_engine = df[df['engine-location'] == 'rear']
print("Rear engine car count:", rear_engine.shape[0])

# Cars with high price (top 10%)
high_price_cutoff = df['price'].quantile(0.9)
premium_cars = df[df['price'] >= high_price_cutoff]
print("Premium car count (top 10%):", premium_cars.shape[0])
````

````python
# Mean price by fuel type
print("Average price by fuel type:\n")
print(df.groupby('fuel-type', observed=False)['price'].mean())

# Mean price by drive-wheels & body-style
print("\nPrice by drive-wheels and body-style:\n")
print(df.groupby(['drive-wheels', 'body-style'], observed=False)['price'].mean())

# Price by engine-location
print("\nPrice by engine location:\n")
print(df.groupby('engine-location', observed=False)['price'].describe())

# Mean price by risk category
print("\nPrice by risk category:\n")
print(df.groupby('risk-category', observed=False)['price'].agg(['mean', 'count']))
````

````python
# Price per unit weight
df['price_per_weight'] = df['price'] / df['curb-weight']
print("Average price per unit curb weight (value per kg):")
print(df['price_per_weight'].describe())

# ratio: width to height
df['width_to_height'] = df['width'] / df['height']
print("\nCar width-to-height ratio distribution:")
print(df['width_to_height'].describe())
````

````python
# price per horsepower
df['price_per_hp'] = df['price'] / df['horsepower']
print("Price per Horsepower calculated.")
print(df['price_per_hp'].head())
````

````python
# Create a feature: performance segment
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

def engine_segment(row):
    if row['horsepower'] <= 100:
        return 'Low Power'
    elif 100 < row['horsepower'] <= 150:
        return 'Moderate Power'
    else:
        return 'High Power'

df['performance-segment'] = df.apply(engine_segment, axis=1).astype('category')

print("Horsepower segment distribution:")
print(df['performance-segment'].value_counts())

````

````python
# Categorizing vehicles based on price into economy, mid-range, and premium
def classify_price(price):
    if price < 10000:
        return 'Economy'
    elif 10000 <= price < 20000:
        return 'Mid-Range'
    else:
        return 'Premium'

df['price_category'] = df['price'].apply(classify_price)
print("Price category counts:")
print(df['price_category'].value_counts())
````

````python
# Average price by make (brand)
brand_prices = df.groupby('make', observed=False)['price'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
top_brands = brand_prices.head(20).index.tolist()

# Flag top-tier brands
df['is_top_brand'] = df['make'].apply(lambda x: 'Yes' if x in top_brands else 'No')

print("Top 20 brands based on average price:")
print(brand_prices.head(20))
print("\nTop brand flag distribution:")
print(df['is_top_brand'].value_counts())
````

````python
# Average engine-size by fuel type.
df['engine-size'] = pd.to_numeric(df['engine-size'], errors='coerce')

print("Average engine size by fuel type:")
print(df.groupby('fuel-type', observed=False)['engine-size'].mean())

# Flag diesel engines
df['is_diesel'] = df['fuel-type'].apply(lambda x: 1 if x == 'diesel' else 0).astype(int)
print("\nDiesel engine count:")
print(df['is_diesel'].sum())
````

````python
# Analyzing how many vehicles in each price category per body style
print("Vehicle price categories by body-style:")
print(df.groupby('body-style')['price_category'].value_counts())
````

````python
# Comparing fuel efficiency vs price
df['avg_mpg'] = (df['city-mpg'] + df['highway-mpg']) / 2
df['price_per_mpg'] = df['price'] / df['avg_mpg']

print("Price per average MPG (miles per gallon):")
print(df['price_per_mpg'].describe())
print("\nTop 5 most expensive cars per MPG (less fuel cost-efficient):")
print(df[['make', 'price_per_mpg']].sort_values(by='price_per_mpg', ascending=False).head(5))

````

````python
# Function to extract statistical summary for any 'make'
def describe_make(make):
    if make in df['make'].unique():
        result = df[df['make'] == make][['price', 'horsepower', 'engine-size', 'curb-weight']].describe()
        print(f"\nStats for '{make}':")
        print(result)
    else:
        print(f"Make '{make}' not found.")

describe_make('jaguar')
describe_make('toyota')

````

````python
#  Difference between City and Highway MPG
df['mpg_gap'] = df['highway-mpg'] - df['city-mpg']

print("Difference between highway and city MPG :\n")
print(df['mpg_gap'].describe())
print("\nCars with the highest highway-to-city mileage gap:\n")
print(df[['make', 'city-mpg', 'highway-mpg', 'mpg_gap']].sort_values(by='mpg_gap', ascending=False).head())
````
